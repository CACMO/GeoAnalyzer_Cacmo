```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar archivo CSV
file_path = "/content/ANTES1DTA (1).CSV"  # Cambia a tu archivo específico
df = pd.read_csv(file_path)

# Mostrar resumen inicial de los datos
print(df.info())
print(df.head())

# Explorar datos faltantes
print("Valores faltantes por columna:")
print(df.isnull().sum())
```

```python
# Estadísticas descriptivas para todas las pruebas
# Convertir todas las columnas (excepto la primera) a numérico, forzando errores a NaN
# The first column 'DTA File Name' appears to be an identifier.
# Convert other columns to numeric, coercing errors to NaN.
for col in df.columns[1:]: # Iterate over all columns from the second one onwards
    df[col] = pd.to_numeric(df[col], errors='coerce')

estadisticas = df.describe()

# Usamos rango intercuartílico (IQR) para detectar valores fuera de rango
# Ensure we operate only on numeric columns for quantile calculations
df_numeric = df.select_dtypes(include=np.number)

q1 = df_numeric.quantile(0.25)
q3 = df_numeric.quantile(0.75)
iqr = q3 - q1

# Valores considerados atípicos
# Apply the outlier detection logic to the numeric subset
atipicos = ((df_numeric < (q1 - 1.5 * iqr)) | (df_numeric > (q3 + 1.5 * iqr)))

# Resumen de valores detectados como anómalos
valores_atipicos = atipicos.sum()
print("Número de valores atípicos detectados por cada prueba:")
print(valores_atipicos)
```
